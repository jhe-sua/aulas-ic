7.Crie um vetor (array unidimensional) cujos valores sejam o resultado da função $f(x) = 3x^2 - 2x +7$ para cada valor de x em um vetor dado com 1000 elementos, construído com linspace.


```python
np.empty((3,3)) #cria uma matriz alocada em qualquer local da maemoria com o tamanho desejado
```

```python
%time
intervalo = np.linspace(-10,10,1000)
funcao = 3*intervalo**2 - 2*intervalo + 7
```

```python
%time
r = np.empty((1000,)) #criase um vetor com uma tupla de um unico elemento
for i in range(len(intervalo)): #com a quantidade de elementos 
    r[i] = 3*intervalo[i]**2 - 2*intervalo[i] + 7
```

```python
plt.plot(intervalo, funcao, "+")
```

```python
plt.plot(intervalo[:-1], np.diff(funcao))
```

```python
plt.plot(intervalo, np.cumsum(funcao))
```
