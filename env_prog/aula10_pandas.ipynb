```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

Criando um DataFrame com numpy

abaixo criamos uma matriz de numeros 1s com numpy, essa matriz foi passada para o pandas, com isso ele cria um dataframe e passamos os nomes das colunas atrave de um list compression

```python
data_frame = pd.DataFrame(np.ones((400,4)), columns=[f"col{numero}" for numero in range(1,5)])
```

Criando um DataFrame com um diccionario, onde cada chave é uma coluna e cada valor é uma lista

```python
data_frame2 = pd.DataFrame({f"col{numero}": range(400) for numero in range(1,5)})
data_frame2
```

Renomeando as colunas de um dataframe

```python
data_frame2.columns = ["x","y","w","z"]
data_frame2
```

```python
data_frame2.values
```

Formas de somar linhas e colunas

```python
data_frame2["v"] = data_frame2[["x","y","w","z"]].sum(axis=1) #axis 1 indica, fazer a operação linha a linha
data_frame2["v2"] = data_frame2["x"] + data_frame2["y"] + data_frame2["w"] + data_frame2["z"] #quando passamos apenas o index um dataframe retorna sua lista de valore
data_frame2
```

```python
#calculando a media por linha
media_por_linha = data_frame2.mean()
media_por_coluna = data_frame2.mean(axis=1)
media_por_coluna
```

Criando um data frame com valores aletorios
o modulo random possui as seguintes funções:

- random = numeros aleatorios entre 0 e 1 random.random(shape)
- ranint = numeros inteiros aletorios entre um intervalo random.randint(inicio, fim, shape)
- uniform = numeros float em um intervalo random.uniform(inicio, fim, shape)

```python
data_frame3 = pd.DataFrame(np.random.randint(1, 100, (400,4)), columns=[f"col{numero}" for numero in range(1,5)])
data_frame3.columns = ["x", "y","z","v"]
data_frame3.describe()
```

Assim como posso transpor um array posso transpor (trocar linhas por colunas e colunasd por linhas) um DataFrame

```python
data_frame3.T
```

fazendo seleções com loc e iloc

```python
data_frame3.loc[(data_frame3["x"] <= 50) & (data_frame3["y"] >= 50)]
#loc posse acessar a linhas e colunas
```

```python
data_frame3[(data_frame3["x"] <= 50) & (data_frame3["y"] >= 50)]
#sem loc acessamos apenas as linhas
```

```python
data_frame3["q1"] = data_frame3.y<=25
data_frame3
```

somando todos os valores que possuem true e todos os que possuem false

```python
data_frame3.groupby("q1").sum().T.sum()
```

Objetos groupBy

```python
g = data_frame3.groupby("q1") #agrupe pelos valores que estao na coluna q1
print(g)
data_frame3.columns
```

com o metodo agg podemos passar um diccionario e decir o que faremos com cada agrupamento

```python
g.agg({"x":"sum","y":"min","z":"max","v":"mean"})
```
