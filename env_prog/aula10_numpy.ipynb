# Aula 10
O índice de massa corporal $IMC=peso/altura^2$ é um critério da Organização Mundial de Saúde para dar uma indicação sobre a condição de peso de uma pessoa adulta. construa uma array com 100 os valores de altura entre 150cm a 195cm, e  100 valores de peso entre 50kg e 100kg. Calcule o IMC para todas as valores e construa uma matriz 2D com os valores de IMC. 

```python
altura = np.linspace(1.50,1.95,100) #criamos um intervalo de 100 alturas espaçadas igualmente "linspace"
peso = np.linspace(50,150,100) #criamos un intervalo de 100 pesos espaçados igualmente 
grid_altura , grid_peso = np.meshgrid(altura, peso) #criamos uma grid com altura e peso, deixando cada uma das alturas em cada coluna e cada um dos pesos em cada linha
```

```python
print(grid_altura) #cada altura em cada coluna
```

```python
print(grid_peso) # cada peso em cada linbha
```

### Indice de massa corporal

$$ IMC = Peso / Altura^2 $$

agora cada um dos pesos sera dividido por cada uma das alturas ao quadrado, pois os pesos se repetem ao redor das linhas e as alturas ao redor das colunas 


```python
%%time
imc = grid_peso/grid_altura**2
print(imc)
```

Calculando a mesma tabela, somente que com for ao inves de utilizar operações matriciais, o tempo é mais demorado

```python
%%time
imc = np.zeros((100, 100))
for altura_inx in enumerate(altura):
    for peso_inx in enumerate(peso):
        imc[altura_inx[0], peso_inx[0]] = peso_inx[1]/altura_inx[1]**2

print(imc)
```

```python
plt.pcolor(imc, cmap="hot", edgecolors="k", linewidths=0.1)
plt.colorbar()

plt.xlabel("")
plt.title("Mapa de calor, imc")
plt.show()
```
